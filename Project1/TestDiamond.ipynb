#Desicion Tree Model

```python
model = '/content/drive/MyDrive/PredictiveAnalytics/Projects/Project1/dtm_file'
```

```python
with open(model, 'rb') as file:
    model = pickle.load(file)
```

```python
#creating a vector - a model expects one row 7 times
x = np.zeros((1, 9))  
 
```

```python
#now assigning values to the test value x by grabing randomly from the data
x[0, 0] = .21
x[0, 1] = 4.00
x[0, 2] = 2.00
x[0, 3] = 2.00
x[0, 4] = 59.80
x[0, 5] = 61.00
x[0, 6] = 3.89
x[0, 7] = 3.84
x[0, 8] = 2.31
x
```

```python
#x = variables/features 'carat', 'cut', 'color', 'clarity', 'depth', 'table', 'x', 'y', 'z'
features = {'carat' : x[0, 0], 'cut' : x[0, 1], 'color' : x[0, 2], 'clarity' : x[0, 3], 'depth' : x[0, 4], 'table' : x[0, 5], 'x' : x[0, 6], 'y' : x[0, 7], 'z' : x[0, 8]}
```

```python
features
```

```python
#Desicion Tree Model Prediction
#now a prediction ..seeing what the Price is based on these features...
prediction = model.predict(x)
print(f'The model predicts that the price is {round(float(prediction[0]), 2)} given these features:\n\t{features}')
```
